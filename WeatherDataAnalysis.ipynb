```python
import pandas as pd
```

```python
# Load the dataset (replace 'your_dataset.csv' with the actual file name)
df = pd.read_csv('WeatherData.csv')

# 1. Find all records where the weather was exactly clear.
clear_weather_records = df[df['Weather'] == 'Clear']

# 2. Find the number of times the wind speed was exactly 4 km/hr.
wind_speed_4_count = df[df['Wind Speed_km/h'] == 4].shape[0]

# 3. Check if there are any NULL values present in the dataset.
null_values = df.isnull().sum()

# 4. Rename the column "Weather" to "Weather_Condition."
df.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)

# 5. What is the mean visibility of the dataset?
mean_visibility = df['Visibility_km'].mean()

# 6. Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
wind_above_24_visibility_25 = df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)].shape[0]

# 7. What is the mean value of each column for each weather condition?
numeric_columns = df.select_dtypes(include='number').columns
mean_values_per_weather_condition = df.groupby('Weather_Condition')[numeric_columns].mean()

# 8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
specific_conditions = df[(df['Weather_Condition'] == 'Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40)]

# 9. Find the number of weather conditions that include snow.
snow_conditions = df['Weather_Condition'].str.contains('Snow', case=False, na=False).sum()
```

```python
# Display results (optional)
print("Clear weather records:\n", clear_weather_records)
print("\nNumber of times wind speed was exactly 4 km/hr:", wind_speed_4_count)
print("\nCheck for NULL values:\n", null_values)
print("\nMean visibility of the dataset:", mean_visibility)
print("\nNumber of records where wind speed > 24 km/hr and visibility = 25 km:", wind_above_24_visibility_25)
print("\nMean value of each column for each weather condition:\n", mean_values_per_weather_condition)
print("\nInstances where the weather is clear, and the relative humidity is > 50, or visibility is > 40:\n", specific_conditions)
print("\nNumber of weather conditions that include snow:", snow_conditions)
```
